```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from tqdm import tqdm 
```

```python
def make_option(gen='random', name_set='us', country='us'):
    return ''.join(['/gen-', gen,'-', name_set,'-', country,'.php'])
```

```python
with open('url.txt', 'r') as file:
      url = file.read()

headers = {'User-Agent': 'Mozilla/5.0'}
DELAY = 0.5
COUNT_PERSONS = 100
```

```python
r = requests.get(url, headers=headers)
html = r.text
soup = BeautifulSoup(html, "html.parser")
```

```python
list_gen_value = [x['value'] for x in soup.find('select', id="gen").find_all('option')]
list_gen_text = [x.text for x in soup.find('select', id="gen").find_all('option')]
dict_gen = dict(zip(list_gen_text, list_gen_value))
list_nameset_value = [x['value'] for x in soup.find('select', id="n").find_all('option')]
list_nameset_text = [x.text for x in soup.find('select', id="n").find_all('option')]
dict_nameset = dict(zip(list_nameset_text, list_nameset_value))
list_country_value = [x['value'] for x in soup.find('select', id="c").find_all('option')]
list_country_text = [x.text for x in soup.find('select', id="c").find_all('option')]
dict_country = dict(zip(list_country_text, list_country_value))

print('Gender: ' , ', '.join(dict_gen.keys()), '\n')
print('Name set: ' , ', '.join(dict_nameset.keys()), '\n')
print('Country: ' , ', '.join(dict_country.keys()), '\n')

url = url + make_option(dict_gen['Random'], dict_nameset['American'], dict_country['United States'])
```

```python
list_personality = []

for t in tqdm(range(COUNT_PERSONS)):
    
    r = requests.get(url, headers=headers)
    html = r.text
    soup = BeautifulSoup(html, "html.parser")
    
    identity = []
    for x in soup.find_all('div', class_="adtl"):
        x.decompose()
        
    identity.append(soup.find('div', class_='address').h3.text)
    identity.append(soup.find('div', class_='adr').text.rstrip().lstrip())
    identity += [x.text for x in soup.find('div', class_='extra').find_all('dd')]
    
    list_personality.append(identity)
    
    time.sleep(DELAY)
    
columns = ['Name', 'Address'] + [x.text for x in soup.find('div', class_='extra').find_all('dt')]
df = pd.DataFrame(data=list_personality, columns=columns)
df.head(10)
```
